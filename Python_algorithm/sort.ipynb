```python
# 選択ソート

data = [9, 3, 7, 1, 4, 2, 5, 8, 6, 0]
print(data, "元のデータ")

for i in range(0, 9):
    # 最小値の位置を保持する変数
    m = i
    for j in range(i+1, 10):
        
        if data[j] < data[m]:
            # mにjの値を代入
            m = j
            
    # data[i]とdata[m]の値を入れ替える        
    data[i], data[m] = data[m], data[i]
    # ソートの過程を出力
    print(data, i+1)
    
print(data, "ソート後のデータ")
```

```python
# バブルソート(右から左に入れ替えていく)

data = [9, 4, 7, 2, 3, 8, 6, 1, 5, 0]
n = len(data)
print(data, "元のデータ")

for i in range(0, n-1):
    for j in range(n-1, i, -1):
        
        if data[j-1] > data[j]:
            # 左右のデータを入れ替える
            data[j], data[j-1] = data[j-1], data[j]
    # ソートの過程を出力
    print(data, i+1)

print(data, "ソート後のデータ")
```

```python
# 挿入ソート

data = [9, 2, 4, 7, 8, 3, 1, 5, 6, 0]
n = len(data)
print(data, "元のデータ")

for i in range(1, n):
    tmp = data[i]
    j = i
    
    # tmpより大きな値を右にずらす
    while j>0 and data[j-1]>tmp:
        data[j] = data[j-1]
        j = j - 1
    
    # ずらした先頭にtmpの値を挿入する
    data[j] = tmp
    # ソートの過程を出力
    print(data, i)
    
print(data, "ソート後のデータ")
```

```python
# 並び変えるデータを乱数で作成し、挿入ソート

import random

n = 15
data = [0]*15
for i in range(n):
    data[i] = random.randint(1, 99)
print(data, "元のデータ")

for i in range(1, n):
    tmp = data[i]
    j = i
    
    # tmpより大きな値を右にずらす
    while j>0 and data[j-1]>tmp:
        data[j] = data[j-1]
        j = j - 1
    
    # ずらした先頭にtmpの値を挿入する
    data[j] = tmp
    # ソートの過程を出力
    print(data, i)
    
print(data, "ソート後のデータ")
```
