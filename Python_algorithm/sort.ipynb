```python
# 選択ソート

data = [9, 3, 7, 1, 4, 2, 5, 8, 6, 0]
print(data, "元のデータ")

for i in range(0, 9):
    # 最小値の位置を保持する変数
    m = i
    for j in range(i+1, 10):
        
        if data[j] < data[m]:
            # mにjの値を代入
            m = j
            
    # data[i]とdata[m]の値を入れ替える        
    data[i], data[m] = data[m], data[i]
    # ソートの過程を出力
    print(data, i+1)
    
print(data, "ソート後のデータ")
```

```python
# バブルソート(右から左に入れ替えていく)

data = [9, 4, 7, 2, 3, 8, 6, 1, 5, 0]
n = len(data)
print(data, "元のデータ")

for i in range(0, n-1):
    for j in range(n-1, i, -1):
        
        if data[j-1] > data[j]:
            # 左右のデータを入れ替える
            data[j], data[j-1] = data[j-1], data[j]
    # ソートの過程を出力
    print(data, i+1)

print(data, "ソート後のデータ")
```

```python
# 挿入ソート

data = [9, 2, 4, 7, 8, 3, 1, 5, 6, 0]
n = len(data)
print(data, "元のデータ")

for i in range(1, n):
    tmp = data[i]
    j = i
    
    # tmpより大きな値を右にずらす
    while j>0 and data[j-1]>tmp:
        data[j] = data[j-1]
        j = j - 1
    
    # ずらした先頭にtmpの値を挿入する
    data[j] = tmp
    # ソートの過程を出力
    print(data, i)
    
print(data, "ソート後のデータ")
```

```python
# 並び変えるデータを乱数で作成し、挿入ソート

import random

n = 15
data = [0]*15
for i in range(n):
    data[i] = random.randint(1, 99)
print(data, "元のデータ")

for i in range(1, n):
    tmp = data[i]
    j = i
    
    # tmpより大きな値を右にずらす
    while j>0 and data[j-1]>tmp:
        data[j] = data[j-1]
        j = j - 1
    
    # ずらした先頭にtmpの値を挿入する
    data[j] = tmp
    # ソートの過程を出力
    print(data, i)
    
print(data, "ソート後のデータ")
```

```python
# クイックソート

import random

n = 15
data = [0]*15
for i in range(n):
    data[i] = random.randint(1, 99)

# クイックソート
def quick_sort(left, right):
    i = left
    j = right
    # 軸の値を決める
    p = data[(left+right)//2]
    
    while True:
        # 軸以上の値の位置を探す
        while data[i] < p:
            i = i + 1
        # 軸以下の値の位置を探す
        while data[j] > p:
            j = j - 1
            
        # 右からの要素と、左からの要素が交わったとき    
        if i >= j:
            break
        # 軸以上と軸以下の値を交換
        data[i], data[j] = data[j], data[i]
        # 位置をずらす
        i = i + 1
        j = j - 1

        # ソートの過程を出力
        print(data)
        
    # 再帰的に実行
    if left < i-1:
        quick_sort(left, i-1)
    if right > j+1:
        quick_sort(j+1, right)
        
        
print(data, "元のデータ")
quick_sort(0, n-1)
print(data, "ソート後のデータ")
```

```python
# マージソート

a = [1, 3, 4, 7, 8, 9]
b = [0, 2, 5, 6]
na = len(a)
nb = len(b)
# マージ用の配列を定義
c = [0]*(na+nb)
# マージに用いる変数
i = 0
j = 0
p = 0

print(a, "データA")
print(b, "データB")

while i < na and j < nb:
    # a[i]がb[j]以下の値の場合
    if a[i] <= b[j]:
        c[p] = a[i]
        i += 1
        p += 1
    else:
        c[p] = b[j]
        j += 1
        p += 1
    # マージの過程を出力
    print(a, "データA")
    print(b, "データB")
    print(c, "データC")
    
# aの残りのデータをcに入れる
while i < na:
    c[p] = a[i]
    i += 1
    p += 1
    # マージの過程を出力
    print(a, "データA")
    print(c, "データC")

# bの残りのデータをcに入れる
while i < nb:
    c[p] = b[j]
    j += 1
    p += 1
    # マージの過程を出力
    print(b, "データB")
    print(c, "データC")
    
print(c, "マージ後のデータ")
```

```python
# ヒープソート

data = [1, 2, 4, 5, 3, 6, 8]
n = len(data)
print(data, "元のデータ")

for i in range((n-1)//2, -1, -1):
    # 並べる親の位置
    p = i
    # 比べる子の位置
    c = p*2+1
    
    while c < n:
        # 兄弟の右のほうが大きい場合
        if c<n-1 and data[c] < data[c+1]:
            c += 1
        # 親が子以上の場合
        if data[p] >= data[c]:
            break
        # 親と子の位置を入れ替える
        data[p], data[c] = data[c], data[p]
        
        # 子の位置を新たな親とする
        p = c
        # 新たなこの位置をcに代入
        c = p*2+1
        # ソートの過程を出力
        print(data)
        
print(data, "初期ヒープ")
```
